**Maximum Pairwise Product**

TEST 1: Small Numbers

```python
#Python
def MaxProduct(a,n):
  index = 0
  for i in range(1,n):
    if a[i] > a[index]:
      index = i
      #print(index)

  index2 = 0
  for i in range(1,n):
    if a[i] != a[index] and a[i] > a[index2]:
      index2 = i 
      #print(index2)

  result = a[index]*a[index2]

  return result

n = int(input())
a = [int(x) for x in input().split()]

MaxProduct(a,n)
```

```python
#C++
int MaxProductC(vector<int>a,int n){
    index = 0;
    for(i=1;i<n;i++){
        if(a[i] > a[index]){
            index = i;
        }
    }
    for(i=1;i<n;i++){
        if(a[i]!=a[index]&&a[i]>a[index2]){
            index2 = i;
        }
    }
    result = a[index]*a[index2];
    return result
}
```

TEST 2: Large input Numbers

```python
n = int(input())
a = [int(x) for x in input().split()]

MaxProduct(a,n)
```

TEST 3: 20 Times random n in [2,20]

```python
n = 20
a = [random.randint(2,20) for i in range (n)]

print(a)
MaxProduct(a,n)
```

```python
a = [i for i in range(1,200001)]
print(MaxProduct(a,200000))
```

**Square String**

```python
#Python
i = int(input())

for i in range(0,i,1):
  par = 0
  word=str(input())
  for i in word:
    par += 1
  split=int((par/2))
  a=word[:split]
  b=word[split:par]
  if par%2 ==0 and a == b: 
      print("yes")
  else:
      print("NO")
```
