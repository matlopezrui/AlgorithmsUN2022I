```python
import numpy as np
import itertools

def test(w,c,items):
  #tests if a combination of items, with weight w fits in a knapsack of capacity c
    return np.sum(np.multiply(w,items)) <= c
```

**1. Count and calculate the exact proportion of Knapsack solutions. for the problem in the image.**

```python
w = [1,1,2,4,12]
c = 15
sol = 0
p = 0

tests = list(itertools.product([0,1], repeat=len(w)))
 
for i in list(tests):
    if test(w,c,np.array(i)):
        sol+=1

p = sol/(2**len(w))
print('Nmero de Soluciones: '+str(sol))
print('Proporciones: '+str(p))
print('Nmero de Combinaciones: '+str(2**len(w)))
```

**2. Estimate the proportion and number of Knapsack solutions. if the sizes are (1,2,3,4,5,6,7,9,10) and the capacity of the knapsack is 10 using Mote Carlo with 1.000, 10.000 random binary vectors.**

```python
w = [i for i in range(11)]
c = 10
sol = 0
p = 0

vectors = 1000
for i in range(1,vectors+1):
  items = np.random.randint(2, size=len(w))
  if test(w,c,items):
      sol+=1

p = sol/vectors
print('Nmero de Soluciones: '+str(sol))
print('Proporciones: '+str(p))
print('Nmero de Combinaciones: ' + str(vectors))
```
