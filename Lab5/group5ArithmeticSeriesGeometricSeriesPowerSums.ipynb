```python
import matplotlib.pyplot as plt

n = 16
a = 15
d = 5
S_n = 0
Serie_S_n = []
Secuencia_a_n = []
for i in range (1, n + 1):
  a_n = a + d * (i - 1)
  S_n = S_n + a_n
  Serie_S_n.append(S_n)
  Secuencia_a_n.append(a_n)

plt.plot(Serie_S_n, color = 'black', label = 'Serie_S_n')
plt.plot(Secuencia_a_n, color = 'green', label = 'Secuencia_a_n')
plt.legend()
```

Geometric Series

```python
n = 20
r = 2
a = 1
S_n = 0
Serie_S_n = []
Secuencia_a_n = []
for i in range (1, n + 1):
  a = a * r
  S_n = S_n + a
  Serie_S_n.append(S_n)
  Secuencia_a_n.append(a)

plt.plot(Serie_S_n, color = 'black', label = 'Serie_S_n')
plt.plot(Secuencia_a_n, color = 'green', label = 'Secuencia_a_n')
plt.legend()
```

Power Sums

```python
n = 6
p = 2
a = 0
S_n = 0
Serie_S_n = []
Secuencia_a_n = []
for i in range (1, n + 1):
 
  a = (i) ** p
  S_n = S_n + a
  Serie_S_n.append(S_n)
  Secuencia_a_n.append(a)
plt.plot(Serie_S_n, color = 'black', label = 'Serie_S_n')
plt.plot(Secuencia_a_n, color = 'green', label = 'Secuencia_a_n')
plt.legend()
```
