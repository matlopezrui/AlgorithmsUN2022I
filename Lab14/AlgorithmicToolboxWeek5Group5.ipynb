**Change Money**

```python
def countMoney(ans, m, n):
  if n == 0:
    return m
  elif n == 1:
    ans.setdefault(1,1)
    return m+1
  elif n == 3:
    ans.setdefault(3,1)
    return m+1
  elif n == 4:
    ans.setdefault(4,1)
    return m+1
  else:
    if n == 6:
      ans.setdefault(6,2)
      return m+2
    elif n == 2:
      ans.setdefault(2,2)
      return m+2
    elif n > 4:
      ans.setdefault(n,m+1)
      return countMoney(ans, m+1, n-4)
  
def changeMoney(n):
  ans = {}
  return countMoney(ans, 0, n)
```

**Primitive Calculator**

```python
def primitiveCalculator(m,n):
  ans[0] = 0
  ans[1] = 1
  if(ans[n] == None):
    print(n, end = " ")
    if (n == 10):
      ans[10] = primitiveCalculator(0,n-1)
      return primitiveCalculator(m+1,int(n/3))
    if((n % 3) == 0):
      ans[int(n/3)] = primitiveCalculator(0,int(n/3))
      return primitiveCalculator(m+1,int(n/3))
    elif((n % 2) == 0):
      ans[int(n/2)] = primitiveCalculator(0,int(n/2))
      return primitiveCalculator(m+1,int(n/2))
    else:
      ans[n-1] = primitiveCalculator(0,n-1)
      return primitiveCalculator(m+1,n-1)
  else:
    return ans[n] + m
```

**Edit Distance**

```python
def editDistance(str1, str2):
  words = [["."] * (len(str2)+1) for i in range(len(str1)+1)]

  for j in range(len(str2) + 1):
    words[len(str1)][j] = len(str2) - j
  for i in range(len(str1) + 1):
    words[i][len(str2)] = len(str1) - i

  for i in range(len(str1) - 1, -1, -1):
    for j in range(len(str2) -1, -1, -1):
      if str1[i]==str2[j]:
        words[i][j] = words[i+1][j+1]
      else:
        words[i][j] = 1 + min(words[i+1][j], words[i][j+1], words[i+1][j+1])

  return words[0][0]
```

```python
a = "short"
b = "ports"
editDistance(a,b)
```

**Longest Common Sequence Two Sequences**

```python
def longestCommonSubsequence(n, a, m, b):
  dp = [[0 for j in range(m+1)] for i in range (n+1)]
  
  for i in range(n-1,-1,-1):
    for j in range(m-1,-1,-1):
      if a[i] == b[j]:
        dp[i][j] = 1 + dp[i+1][j+1]
      else:
        dp[i][j] = max(dp[i][j+1], dp[i+1][j])
  return dp[0][0]
```

```python
n = 3
a = [2,7,5]
m = 2
b = [2,5]
print(longestCommonSubsequence(n, a, m, b))

n = 1
a = [7]
m = 4
b = [1,2,3,4]
print(longestCommonSubsequence(n, a, m, b))

n = 4
a = [2,7,8,3]
m = 4
b = [5,2,8,7]
print(longestCommonSubsequence(n, a, m, b))
```

```python
def longestCommonSubsequenceThree(n, a, m, b, l, c):
  dp = [[[0 for k in range(l+1)] for j in range (m+1)]for i in range(n+1)]
  
  for i in range(n-1,-1,-1):
    for j in range(m-1,-1,-1):
      for k in range(l-1,-1,-1):
        if a[i] == b[j] and b[j] == c[k]:
          dp[i][j][k] = 1 + dp[i+1][j+1][k+1]
        else:
          dp[i][j][k] = max(dp[i][j+1][k], dp[i+1][j][k], dp[i][j][k+1])
  return dp[0][0][0]
```

```python
n = 3
a = [1,2,3]
m = 3
b = [2,1,3]
l = 3
c = [1,3,5]
print(longestCommonSubsequenceThree(n, a, m, b, l, c))

n = 5
a = [8,3,2,1,7]
m = 7
b = [8,2,1,3,8,10,7]
l = 6
c = [6,8,3,1,4,7]
print(longestCommonSubsequenceThree(n, a, m, b, l, c))
```
