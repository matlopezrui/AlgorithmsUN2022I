**Fibonacci Number**

```python
def fibo(n):
  f = [None]*(n+1)
  if n == 0:
    return 0
  elif n == 1:
    return 1;
  elif n >= 2:
    f[0] = 0
    f[1] = 1
    for i in range(2,n+1):
      f[i] = f[i-1] + f[i-2]
  return f[n]
```

```python
n = int(input())
fibo(n)
```

**Last Digit of a Large Fibonacci Number**

```python
def fibo_last(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  f = [None]*(n)
  f[0] = 1
  f[1] = 1
  for i in range(2,n):
    f[i] = (f[i-1]+f[i-2])%10
  return f[n-1]
```

```python
n = int(input())
print(fibo_last(n))
```

**Greatest Common Divisor**

```python
def gcd(a,b):
     
  if (a == 0):
    return b
  if (b == 0):
    return a

  if (a == b):
    return a
        
  if (a > b):
    return gcd(a%b, b)
  else:
    return gcd(a, b%a)
```

```python
a, b = map(int, input().split())
gcd(a,b)
```

**Least Common Multiple**

```python
def lcm(a, b):
  return (a / gcd(a,b))* b
```

```python
a, b = map(int, input().split())
print(lcm(a, b))
```

**Fibonacci Number Again (Pisano Period)**

```python
def pisano(m):
  previous, current = 0, 1
  for i in range(0, m * m):
    previous, current = current, (previous + current) % m

    if (previous == 0 and current == 1):
      return i + 1
```

```python
def get_fibonacci_huge(n, m):

  p = pisano(m)

  n = n % p

  return (fibo(n) % m)
```

```python
n, m = map(int, input().split())
print(get_fibonacci_huge(n, m))
```

**Last DIgit of the Sum of Fibonacci Numbers**

```python
def fibonacci_sum(n):
  if n == 1:
    return 1
  elif n == 2:
    return 2

  current = 2
  sum = 2

  for i in range(2,n):
    if i >= 3:
      current = get_fibonacci_huge(i+1,10)
    sum += current
    sum = sum%10

  return sum
```

```python
n = int(input())
print(fibonacci_sum(n))
```

**Last Digit of the Sum of Fibonacci Numbers Again**

```python
def fibonacci_partial_sum(n, m):
  if n > m:
    return "Error"

  sum = 0

  for i in range(n, m + 1):
    current = get_fibonacci_huge(i,10)
    sum += current
    sum = sum%10

  return sum
```

```python
n, m = map(int, input().split())
print(fibonacci_partial_sum(n, m))
```

**Last Digit of The Sum of Squares of Fibonacci Numbers**


```python
def fibonacci_sum_squares_naive(n):
  if n <= 1:
    return n

  sum = get_fibonacci_huge(n,10)*get_fibonacci_huge(n+1,10)
  sum = sum%10

  return sum
```

```python
n = int(input())
print(fibonacci_sum_squares_naive(n))
```
